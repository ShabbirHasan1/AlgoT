```python
# butterfly spread strategy for 2018-01-01 with expiry date 2018-01-04
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
premium = 5
def call_payoff(spot_price, strike_price, premium):
    return np.where(spot_price > strike_price, spot_price - strike_price, 0) - premium  
```

```python
std_dev = merged_df_2018['Strike Price'].std()
higher_strike = merged_df_2018['Strike Price'].mean() + std_dev
lower_strike = merged_df_2018['Strike Price'].mean() - std_dev
short_call = merged_df_2018['Strike Price'].min() + std_dev
```

```python
spot_prices = merged_df['Spot Price'].values
lower_strike_long_call_payoffs = call_payoff(spot_prices, lower_strike, premium)
higher_strike_long_call_payoffs = call_payoff(spot_prices, higher_strike, premium)
short_call_payoffs = -call_payoff(spot_prices, short_call, premium)
```
