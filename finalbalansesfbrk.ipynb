```python
# Convert the Datetime column to datetime format
data['Datetime'] = pd.to_datetime(data['Datetime'], errors='coerce')

# Extract rows where the time is 15:15 (3:15 PM)
data_1515 = data[data['Datetime'].dt.time == pd.Timestamp('15:15:00').time()]

# Create a new dataframe with Date and cumReturns at 3:15 PM
daily_balance = data_1515[['Datetime', 'cumReturns']].copy()

# Extract only the date part for clarity
daily_balance['Date'] = daily_balance['Datetime'].dt.date

# Drop the original Datetime column
daily_balance.drop(columns=['Datetime'], inplace=True)

# Rename columns for clarity
daily_balance.rename(columns={'cumReturns': 'CumReturns'}, inplace=True)

# Convert the Date column to datetime format in brokerage data for merging
brokerage_data['Date'] = pd.to_datetime(brokerage_data['Date']).dt.date

# Merge the daily balance dataframe with the brokerage data on Date
merged_data = pd.merge(daily_balance, brokerage_data[['Date', 'Daily Brokerage']], on='Date', how='left')

# Initialize the Final Balance column with the initial balance of 100000
initial_balance = 100000
merged_data['Final Balance'] = initial_balance

# Fill NaN values in CumReturns with 0
merged_data['CumReturns'] = merged_data['CumReturns'].fillna(0)

# Calculate the Final Balance based on the condition provided
for i in range(len(merged_data)):
    if i == 0:  # For the first day
        if merged_data.loc[i, 'CumReturns'] != 0:
            merged_data.loc[i, 'Final Balance'] = initial_balance + merged_data.loc[i, 'CumReturns'] - merged_data.loc[i, 'Daily Brokerage']
        else:
            merged_data.loc[i, 'Final Balance'] = initial_balance
    else:  # For subsequent days
        if merged_data.loc[i, 'CumReturns'] != 0:
            merged_data.loc[i, 'Final Balance'] = merged_data.loc[i-1, 'Final Balance'] + merged_data.loc[i, 'CumReturns'] - merged_data.loc[i, 'Daily Brokerage']
        else:
            merged_data.loc[i, 'Final Balance'] = merged_data.loc[i-1, 'Final Balance']


merged_data.head()
```
